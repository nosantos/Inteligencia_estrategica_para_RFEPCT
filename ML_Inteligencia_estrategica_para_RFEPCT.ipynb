##Introdução

A Rede Federal de Educação Profissional, Científica e Tecnológica (RFEPCT), criada pela [Lei nº 11.892/2008](https://www.google.com/url?q=https%3A%2F%2Fwww.planalto.gov.br%2Fccivil_03%2F_ato2007-2010%2F2008%2Flei%2Fl11892.htm), possui a missão estratégica de promover a educação profissional alinhada ao desenvolvimento socioeconômico local e regional. Com a recente expansão anunciada pelo Governo Federal em 2024, que prevê a criação de 100 novos *campi*, o desafio de garantir a pertinência social da oferta educacional torna-se ainda mais crítico.

Historicamente, o planejamento de novos cursos enfrenta dificuldades em correlacionar, de forma sistemática e escalável, a matriz de oferta acadêmica com a realidade econômica dos territórios. O desalinhamento entre os cursos ofertados e os arranjos produtivos locais (APLs) pode resultar em baixas taxas de empregabilidade para os egressos e na subutilização da infraestrutura instalada, comprometendo o papel dos Institutos Federais como vetores de desenvolvimento regional.

Este projeto de **Machine Learning** surge para abordar a dimensão territorial e identificar "Onde o curso é necessário". Utilizando algoritmos de aprendizado não supervisionado (*Clustering*), propõe-se uma metodologia baseada em dados para agrupar os 5.570 municípios brasileiros segundo suas vocações econômicas reais, identificadas a partir de indicadores de PIB, emprego formal e dinâmica setorial.

O objetivo central é transcender as divisões geográficas administrativas (Estados e Regiões) e identificar "territórios econômicos" semelhantes. Ao cruzar esses perfis com a oferta atual de cursos, torna-se possível:

1. **Diagnosticar** a aderência da oferta atual à vocação econômica do município;
2. **Identificar** demandas reprimidas em regiões sem cobertura;
3. **Orientar** a abertura de novos eixos tecnológicos com base em evidências estatísticas, e não apenas em tendências históricas.

Dessa forma, a aplicação do modelo de Clustering apoioa os gestores na tomada de decisão, assegurando que a expansão da Rede Federal cumpra seu propósito de interiorização e desenvolvimento sustentável.

O desenvolvimento deste projeto de Machine Learning foi realizado considerando as seguintes etapas estruturantes: **definição do problema, planejamento da solução, implementação e elaboração da documentação.**

**Implementação:**
A implementação do modelo de Clustering seguiu um *pipeline* de processamento, descrito nas seguintes etapas:

1. **Preparação do Ambiente e Dados:** Instalação das bibliotecas analíticas (*Pandas, Scikit-Learn, Seaborn*) e ingestão dos dados brutos das fontes (IBGE, CAGED e PNP), seguida pela unificação dos *datasets* através do código IBGE do município.
2. **Pré-processamento:** Tratamento de valores nulos e, crucialmente, a aplicação de técnicas de normalização (*StandardScaler*) para padronizar as escalas das variáveis numéricas (ex: equiparar a magnitude do PIB com a de saldos de emprego), evitando vieses no cálculo de distâncias do algoritmo.

Nesta etapa, também foi realizado o ajuste fino de hiperparâmetros (*Grid Search*) para garantir a estabilidade do modelo.
4. **Explicabilidade e Validação:** Treinamento de um modelo substituto de Árvore de Decisão para extrair as regras de negócio que definiram cada grupo, permitindo a "tradução" dos clusters matemáticos em perfis econômicos inteligíveis (ex: "Polo Agropecuário", "Centro Econômico Diversificado").
5. **Aplicação Estratégica:** Cruzamento final dos clusters identificados com os dados da Plataforma Nilo Peçanha para gerar os diagnósticos de aderência e os mapas de oportunidades.

**Limitações atuais e futuras**
O desenvolvimento de um modelo de agrupamento territorial, embora represente um avanço significativo em relação ao planejamento tradicional, apresenta limitações inerentes à técnica e à disponibilidade de dados:

* **Defasagem Temporal dos Dados (Limitação Atual):** O modelo depende da atualização das bases públicas (IBGE/CAGED). O "atraso" natural desses indicadores pode fazer com que mudanças econômicas muito recentes (ex: instalação de uma nova fábrica em 2025) ainda não se reflitam na classificação do cluster.
* **Rigidez do Agrupamento (Limitação Atual):** O algoritmo K-Means atribui cada município a um único cluster predominante. Cidades com vocações híbridas (ex: forte em Agro e Turismo simultaneamente) podem ter sua complexidade simplificada pelo modelo.
* **Integração Sistêmica (Perspectiva Futura):** Atualmente, a solução roda em ambiente de desenvolvimento (Notebook). A evolução natural é a integração via API com os sistemas de gestão acadêmica (Ex.: SUAP), permitindo que o alerta de vocação apareça automaticamente no momento do cadastro de um novo eixo.
* **Evolução para Modelos Probabilísticos (Perspectiva Futura):** Futuras versões podem utilizar algoritmos de *Soft Clustering* (como GMM), que em vez de uma classificação única, indicariam probabilidades (ex: "Esta cidade é 70% Agro e 30% Serviços"), refinando ainda mais a recomendação pedagógica.

##**Descrição do problema**

* **Objetivo do modelo de ML:** Segmentar os 5.570 municípios brasileiros em grupos homogêneos (*clusters*) baseados em variáveis socioeconômicas (PIB, mercado de trabalho e dinâmica setorial), permitindo a identificação automática de vocações territoriais.
* **Enquadramento do problema:** Aprendizado de Máquina Não Supervisionado (*Unsupervised Learning*), especificamente uma tarefa de Agrupamento (*Clustering*), dada a ausência de rótulos prévios sobre a classificação econômica das cidades.
* **Tipos de modelo ML testados:**
  * **K-Means:** Algoritmo baseado em centróides (selecionado como modelo final).
  * **Clusterização Hierárquica (*Agglomerative Clustering*):** Abordagem baseada em conectividade.
  * **Gaussian Mixture Models (GMM):** Abordagem probabilística baseada em densidade.


* **Medida de desempenho:** A avaliação da qualidade dos agrupamentos e a comparação entre os modelos basearam-se nas seguintes métricas de validação interna:
  * **Silhouette Score (Maior é melhor):** Mede a coesão e a separação dos clusters (intervalo de -1 a 1).
  * **Índice Calinski-Harabasz (Maior é melhor):** Avalia a razão entre a dispersão intra-cluster e inter-cluster.
  * **Índice Davies-Bouldin (Menor é melhor):** Mede a similaridade média entre cada cluster e seu cluster mais semelhante.

- **Estratégia de Análise de Correlação e Agrupamento de Variáveis**

Visando otimizar a seleção de características (*feature selection*) e promover a redução de dimensionalidade sem perda de informação relevante, as variáveis são segmentadas em **quatro grupos lógicos**. Esta divisão baseia-se na natureza intrínseca dos dados (Riqueza, Escala, Desempenho e Qualidade), permitindo uma análise de correlação intragrupo mais assertiva e tecnicamente justificada.

Abaixo, detalham-se os grupos definidos e a racionalidade analítica aplicada:

**Grupo 1: Indicadores Macroeconômicos (Riqueza e Produção)**

* **Objetivo:** Identificar e tratar a redundância estatística entre o Produto Interno Bruto (PIB) total e os Valores a Preços Corretentes setoriais, visto que, historicamente, o Valor Preços Correntes Total (`vl_total`) tende a apresentar correlação próxima de 1.0 com o PIB (`pib`).
* **Variáveis Analisadas:** `pib` (Produto Interno Bruto), `imposto`, `vl_total` (VAB Total) e os desdobramentos setoriais (`vl_agro`, `vl_ind`, `vl_serv`, `vl_adm`).
* **Critério de Seleção:** Priorizar a variável de maior explicabilidade (PIB) em caso de redundância com o Vl Total, e verificar a independência linear entre os setores produtivos.

**Grupo 2: Indicadores de Escala do Mercado de Trabalho (Volume)**

* **Objetivo:** Analisar a correlação entre o tamanho do mercado de trabalho (`est_mes`) e a movimentação bruta de pessoal (admissões e desligamentos).
* **Justificativa Técnica:** A premissa adotada é a de que, em municípios de maior porte, os volumes absolutos de admissões (`admi`) e desligamentos (`desl`) apresentam alta colinearidade com o estoque total (`est_mes`). Confirmada essa hipótese, torna-se possível reduzir a dimensionalidade mantendo apenas a variável de Estoque como representativa da dimensão "Tamanho da Cidade", eliminando variáveis redundantes que contam a mesma história estatística.
* **Variáveis Analisadas:** Totais (`est_mes_total`, `admi_total`, `desl_total`) e suas respectivas quebras setoriais.

**Grupo 3: Indicadores de Desempenho Líquido (Saldos)**

* **Objetivo:** Avaliar o saldo de empregos (`admi - desl`) como uma métrica de dinâmica econômica, distinta das métricas de escala.
* **Justificativa Técnica:** O saldo é crucial para o *clustering* pois diferencia a "direção" da economia (expansão ou retração). Espera-se que esta variável apresente baixa correlação com o PIB ou Estoque, pois uma cidade pequena pode apresentar saldo proporcionalmente alto (crescimento acelerado), enquanto uma metrópole pode apresentar estagnação.
* **Variáveis Analisadas:** `saldo_total` e os saldos setoriais (`saldo_agro`, `saldo_ind`, `saldo_const`, `saldo_com`, `saldo_serv`).

**Grupo 4: Indicadores Qualitativos e Relativos (Intensidade)**

* **Objetivo:** Analisar métricas que refletem a qualidade e a característica do vínculo empregatício, independentemente do porte populacional do município.
* **Justificativa Técnica:** Variáveis como o tempo médio de emprego (`tp_emp`) e a variação relativa (`vr_rel`) funcionam como indicadores de intensidade/taxa. Elas são essenciais para distinguir perfis econômicos específicos, como cidades com predominância de trabalho sazonal/temporário (ex: safra agrícola) *versus* cidades com vínculos estáveis (ex: serviço público ou indústria de base).
* **Variáveis Analisadas:** Tempo de emprego (`tp_emp_total` e setoriais) e Variação Relativa (`vr_rel_total` e setoriais).

---

**Resumo da Aplicação:**
A aplicação desta estratégia permite conduzir a análise de correlação de forma segmentada, facilitando a decisão de descarte de variáveis redundantes (como `vl_total` e fluxos brutos de `admi`/`desl` quando explicados pelo estoque) e a manutenção de variáveis ortogonais (independentes) essenciais para a qualidade do agrupamento final.

- **Análise de correlação**

```python
# Definindo os grupos de variáveis baseados na lógica aprovada
# Certifique-se de que os nomes das colunas no seu df correspondem a estes (com underscores)

grupos = {
    "1. Macroeconômico (Riqueza)": [
        'pib', 'imposto', 'vl_total',
        'vl_agro', 'vl_ind', 'vl_serv', 'vl_adm'
    ],
    "2. Escala do Mercado (Volume: Estoque vs Movimentação)": [
        'est_mes_total', 'admi_total', 'desl_total',
        'est_mes_agro', 'admi_agro', 'desl_agro',
        'est_mes_ind', 'admi_ind', 'desl_ind',
        'est_mes_serv', 'admi_serv', 'desl_serv',
        'est_mes_com', 'admi_com', 'desl_com',
        'est_mes_const', 'admi_const', 'desl_const'
    ],
    "3. Desempenho Líquido (Saldos)": [
        'saldo_total',
        'saldo_agro', 'saldo_ind', 'saldo_serv',
        'saldo_com', 'saldo_const'
    ],
    "4. Qualitativo e Relativo (Tempo e Variação)": [
        'tp_emp_total', 'vr_rel_total',
        'tp_emp_agro', 'vr_rel_agro',
        'tp_emp_ind', 'vr_rel_ind',
        'tp_emp_serv', 'vr_rel_serv',
        'tp_emp_com', 'vr_rel_com',
        'tp_emp_const', 'vr_rel_const'
    ]
}

df = df_ml

# Loop para gerar um gráfico por grupo
for nome_grupo, colunas in grupos.items():
    # Verifica quais colunas existem no dataframe atual para evitar erros
    cols_existentes = [c for c in colunas if c in df.columns]

    if len(cols_existentes) < 2:
        print(f"⚠️ Grupo '{nome_grupo}' ignorado: menos de 2 colunas encontradas.")
        continue

    plt.figure(figsize=(12, 8))

    # Cálculo da correlação
    corr = df[cols_existentes].corr()

    # Criação da máscara para ocultar a metade superior (triângulo superior)
    # pois é espelhada, facilitando a leitura
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Plot do Heatmap
    sns.heatmap(corr,
                mask=mask,
                annot=True,
                fmt=".2f",
                cmap='coolwarm',
                vmin=-1, vmax=1,
                linewidths=0.5)

    plt.title(f"Matriz de Correlação: {nome_grupo}", fontsize=14, pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("\n" + "="*80 + "\n")
```

O histograma com zoom mostra uma distribuição centrada em zero.


* **Tratamento Necessário:**
  * **QuantileTransformer:** Para as variáveis relativas (`vr_rel`), o `StandardScaler` não funciona bem pois há *outlier* extremo. O `QuantileTransformer` é aplicado pois realiza uma transformação não linear baseada em quantis, forçando os dados a seguirem uma distribuição (como a normal ou uniforme) e reduzindo drasticamente o impacto dos *outliers*.
  * **Para o tempo** (`tp_emp`), o `StandardScaler` ou `Yeo-Johnson` funcionam bem.

---
<br>

**Resumo do Plano de Pré-processamento:**

|Grupo de Variáveis|Característica dos Dados|Tratamento (Pipeline)|
|-|-|-|
|**Riqueza (pib, vl_*)**| Apenas positivos, Cauda Longa extrema| 1. np.log1p (Logaritmo)|
||| 2. StandardScaler|
|**Escala (est_mes_*)**| Apenas positivos, Cauda Longa extrema|1. np.log1p (Logaritmo)|
||| 2. StandardScaler|
|**Saldos (saldo_*)**| Positivos e Negativos, Pico no Zero|1. PowerTransformer(method='yeo-johnson')|
|**Qualidade (tp_emp)**| Leve assimetria, positivos|1. StandardScaler|
|**Variação(vr_rel)**| Outliers extremos, percentuais|1. QuantileTransformer|



```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, QuantileTransformer, PowerTransformer, FunctionTransformer

# Função auxiliar para garantir não-negatividade antes do Log
def clip_negatives(X):
    return np.clip(X, a_min=0, a_max=None)

# ==============================================================================
# 1. DEFINIÇÃO DAS COLUNAS POR GRUPO DE TRATAMENTO
# ==============================================================================

# Grupo 1 & 2: Riqueza e Escala (Assimetria Positiva -> Clip + Log1p + Standard)
cols_log_scale = [
    'pib', 'vl_agro', 'vl_ind', 'vl_adm',
    'est_mes_total', 'est_mes_agro', 'est_mes_ind'
]

# Grupo 3: Saldos (Positivos/Negativos -> Yeo-Johnson)
cols_yeo_johnson = [
    'saldo_agro', 'saldo_ind', 'saldo_com', 'saldo_serv', 'saldo_const'
]

# Grupo 4a: Variação Relativa (Outliers extremos -> QuantileTransformer)
# ALTERAÇÃO: Trocamos para QuantileTransformer para forçar normalidade nos outliers
cols_quantile = [
    'vr_rel_total'
]

# Grupo 4b: Tempo de Emprego (Distribuição comportada -> StandardScaler)
cols_standard = [
    'tp_emp_total', 'tp_emp_agro', 'tp_emp_ind'
]

# ==============================================================================
# 2. CONSTRUÇÃO DO PIPELINE DE PRÉ-PROCESSAMENTO (ATUALIZADO)
# ==============================================================================

# Pipeline específico para aplicar Clip -> Log -> Padronização
# ALTERAÇÃO: Adicionado 'clip' para resolver o erro do vl_ind
log_scale_pipeline = Pipeline(steps=[
    ('clip', FunctionTransformer(clip_negatives, validate=False)),
    ('log', FunctionTransformer(np.log1p, validate=False)),
    ('scaler', StandardScaler())
])

# Processador Central
preprocessor = ColumnTransformer(
    transformers=[
        ('log_scale', log_scale_pipeline, cols_log_scale),
        ('yeo_johnson', PowerTransformer(method='yeo-johnson'), cols_yeo_johnson),
        ('quantile', QuantileTransformer(output_distribution='normal', random_state=42), cols_quantile),
        ('standard', StandardScaler(), cols_standard)
    ],
    verbose_feature_names_out=False
)

# ==============================================================================
# 3. EXECUÇÃO DA TRANSFORMAÇÃO E VALIDAÇÃO
# ==============================================================================

# Ajusta e transforma os dados
# Usamos fillna(0) no DF original preventivamente para garantir integridade,
# mas o pipeline já trataria a maioria dos casos.
df = df_ml_corr # atribui o dataframe para transformação
X_processed = preprocessor.fit_transform(df.fillna(0))

# Recupera os nomes das colunas
features_finais = (cols_log_scale +
                   cols_yeo_johnson +
                   cols_quantile +
                   cols_standard)

# Cria o DataFrame processado final
df_scaled = pd.DataFrame(X_processed, columns=features_finais, index=df.index)

# Visualização do Resultado
print("Pré-processamento (V2) concluído.")
print(f"Shape original: {df.shape}")
print(f"Shape processado: {df_scaled.shape}") # Deve ser (5570, 16) - sem perdas
print("\nPrimeiras 5 linhas dos dados normalizados:")
display(df_scaled.head())

# Estatísticas descritivas para validar
print("\nEstatísticas descritivas (Validação Final):")
# Focamos na validação do desvio padrão (std) e min/max
display(df_scaled.describe().round(2))
```

##**Modelagem**
A etapa de Modelagem representa o núcleo analítico do projeto, onde algoritmos de Aprendizado Não Supervisionado foram aplicados para transformar os dados processados em inteligência territorial. Dada a ausência de rótulos prévios (*ground truth*) sobre a vocação econômica dos municípios, o processo de construção do modelo foi estruturado em duas fases críticas de validação estatística:

* **Definição do número de clusters:** Investigação da quantidade ideal de grupos () que melhor representa a diversidade econômica brasileira, utilizando métodos gráficos e métricas de coesão (Método do Cotovelo e Índice de Silhueta).
* **Comparação entre modelos:** Uma bateria de testes comparativos entre diferentes abordagens algorítmicas — baseada em centróides (**K-Means**), conectividade (**Hierarchical**) e distribuição probabilística (**GMM**) — avaliadas sob a ótica de três indicadores de desempenho rigorosos: **Silhouette Score**, **Calinski-Harabasz** e **Davies-Bouldin**.

### **Definição do número de clustering**
Em algoritmos de aprendizado não supervisionado, a determinação da quantidade ideal de agrupamentos ($k$) é uma etapa crítica, pois o modelo não possui referências prévias de classes. Para superar essa incerteza e evitar a arbitrariedade, é aplicada uma abordagem matemática baseada em dois testes fundamentais de validação:

* **Método do Cotovelo (*Elbow Method*):** Este método avalia a "Inércia" do modelo, ou seja, a compactação interna dos grupos (soma dos quadrados das distâncias intra-cluster). O objetivo analítico é identificar o ponto de inflexão na curva — o "cotovelo" — onde o ganho de compactação se torna marginal ao se adicionar novos grupos, indicando o limite da eficiência estatística.
* **Índice de Silhueta (*Silhouette Score*):** Métrica utilizada para mensurar a qualidade da separação e a coesão dos grupos formados. Variando de -1 a +1, o índice quantifica o quão semelhante um município é ao seu próprio cluster em comparação aos demais. Valores mais próximos de +1 indicam uma estrutura de agrupamento robusta, onde os elementos estão bem posicionados e distantes das fronteiras de decisão vizinhas.

Após a execução destes testes e a análise conjunta dos indicadores gráficos, **definiu-se o número de 7 clusters** como a configuração ideal, pois apresentou o melhor equilíbrio entre compactação (baixa inércia) e distinção (alta silhueta), permitindo uma segmentação econômica detalhada sem gerar fragmentação excessiva.

- **Método do Cotovelo (Elbow Method) e Índice de Silhueta (Silhouette Score)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Configuração
range_n_clusters = range(2, 13) # Testaremos de 2 a 12 grupos
inertias = []
silhouettes = []

print("Iniciando a busca pelo K ideal (isso pode levar alguns segundos)...")

for k in range_n_clusters:
    # Cria o modelo KMeans para K grupos
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

    # Treina com os dados normalizados
    cluster_labels = kmeans.fit_predict(df_scaled)

    # Armazena as métricas
    inertias.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouettes.append(silhouette_avg)

    #print(f"K={k}: Silhueta = {silhouette_avg:.4f}")
    print(f"K = {k:2d} | Silhueta: {silhouette_avg:.4f} | Inércia: {kmeans.inertia_:.2f}")

# ==============================================================================
# VISUALIZAÇÃO DOS RESULTADOS
# ==============================================================================
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico 1: Método do Cotovelo (Inércia)
color = 'tab:blue'
ax1.set_xlabel('Número de Clusters (K)', fontsize=12)
ax1.set_ylabel('Inércia (Soma dos Quadrados Intra-Cluster)', color=color, fontsize=12)
ax1.plot(range_n_clusters, inertias, 'o--', color=color, linewidth=2, markersize=8)
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, alpha=0.3)

# Eixo gêmeo para a Silhueta
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Índice de Silhueta (Qualidade da Separação)', color=color, fontsize=12)
ax2.plot(range_n_clusters, silhouettes, 's-', color=color, linewidth=2, markersize=8)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Definição do K Ideal: Cotovelo vs Silhueta', fontsize=14, pad=20)
plt.xticks(range_n_clusters)
plt.tight_layout()
plt.show()
```

- **Interpretação dos Resultados e Definição do K**



A análise conjunta dos gráficos do Cotovelo (*Elbow*) e da Silhueta revelou um cenário analítico que exige uma interpretação para além da maximização absoluta dos índices estatísticos. A decisão de modelagem baseou-se nos seguintes achados:

- **O Comportamento do K=2 (Limitação da Dicotomia):**
Matematicamente, a divisão em 2 clusters apresentou o maior coeficiente de Silhueta (0,26). Este resultado é frequente em bases de dados socioeconômicos, refletindo a desigualdade estrutural brasileira que divide o país basicamente em "Grandes Centros Econômicos" e "Municípios de Pequeno Porte".
No entanto, para fins de política pública e alocação de cursos técnicos, essa segmentação binária mostrou-se **insuficiente**, pois não captura as nuances vocacionais necessárias (como a distinção entre um polo agropecuário e um polo turístico).
<br>

- **O Pico Local em K=7 (Estabilidade):**
Ao analisar a evolução do coeficiente de Silhueta, observou-se uma tendência de queda entre  ($k=2$) e ($k=5$) . Contudo, houve uma recuperação relevante do índice em ($k=7$) (0,1644), superando os valores de ($k=5$), ($k=6$) e ($k=8$).
Esse fenômeno indica um "pico local" de qualidade, sugerindo que os dados possuem uma aderência natural a uma divisão em sete grupos. Estatisticamente, isso demonstra que, ao fixar 7 clusters, o algoritmo conseguiu formar grupos mais coesos e bem definidos do que nas configurações vizinhas.
<br>

- **Inércia e Compactação:**
Corroborando a análise da Silhueta, a curva do Cotovelo demonstrou que em  ($k=7$) a inércia (variância intra-cluster) já havia sido reduzida significativamente (para níveis inferiores a 50.000), indicando a formação de grupos compactos sem a necessidade de fragmentar excessivamente a base de dados.
<br>

**Decisão Final: Adoção de 7 Clusters:**
Com base nessas evidências, definiu-se a utilização de **7 Clusters** para o modelo final. A escolha justifica-se pelo equilíbrio entre o rigor estatístico (superioridade do índice de Silhueta frente às opções de 5 a 8 grupos) e a **aplicabilidade estratégica**. A granularidade de 7 perfis permite mapear com precisão as diversas realidades econômicas (Industrial, Agro, Serviços, etc.), facilitando o alinhamento direto com os 13 Eixos Tecnológicos da oferta educacional, objetivo central deste trabalho.

### **Comparação entre modelos (K-Means, Hierarchical e GMM)**
Visando garantir a robustez estatística dos resultados e mitigar vieses inerentes ao algoritmo K-Means (que pressupõe clusters com variância esférica e tamanhos similares), é realizada uma análise comparativa com outras famílias de algoritmos. O objetivo desta etapa é verificar se a estrutura dos dados se comportaria melhor sob premissas matemáticas distintas.

**Modelos Adicionais Avaliados**
Além do K-Means, são testados dois modelos robustos de literatura:

* **Clusterização Hierárquica (Aglomerativa):** Diferente da abordagem de particionamento do K-Means, este método constrói uma hierarquia de grupos (dendrograma) baseada na conectividade. Foi selecionado por sua eficácia em identificar como pequenos municípios se aglutinam naturalmente, além de eliminar a necessidade de inicialização aleatória de centróides.
* **Gaussian Mixture Models (GMM):** Uma abordagem probabilística que flexibiliza a premissa geométrica. Enquanto o K-Means assume formas circulares, o GMM assume que os dados são gerados por uma mistura de distribuições gaussianas, permitindo clusters de formato elíptico. Esta característica é crucial em dados socioeconômicos correlacionados (ex: PIB e Estoque de Emprego), onde os grupos tendem a ter formas alongadas.

**Métricas de Avaliação de Desempenho**
Como se trata de aprendizado não supervisionado (sem rótulos verdadeiros para conferência), a comparação entre os modelos baseou-se em métricas de validação interna:

1. **Silhouette Score:** Avalia a coesão e a separação dos grupos. Quanto **maior** o valor, melhor a definição do cluster.
2. **Índice Calinski-Harabasz:** Mede a razão entre a dispersão intra-cluster e inter-cluster. Quanto **maior** o valor, mais distintos são os grupos.
3. **Índice Davies-Bouldin:** Avalia a similaridade média entre cada cluster e seu vizinho mais próximo. Quanto **menor** o valor, melhor a separação (indica menos confusão entre os grupos).

Após comparação entre os resultados das métricas aplicadas o **K-Means (com  $k=7$ )** foi selecionado como o algoritmo definitivo para a geração dos clusters.

- **Modelos e métricas de desempenho**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# ==============================================================================
# 0. BLINDAGEM DOS DADOS (Correção da Variação)
# ==============================================================================
# Seleciona apenas as colunas que NÃO são 'cluster' ou 'Perfil_Economico'
# Isso garante que o modelo treine apenas com as features, independente do estado do df
cols_para_treino = [col for col in df_scaled.columns if col not in ['cluster', 'Perfil_Economico', 'Eixos_Recomendados']]
print(f"Treinando modelos com {len(cols_para_treino)} variáveis (Blindagem Ativa).")

# Configuração
n_clusters = 7
results = []

print(f"Iniciando comparação de modelos com K={n_clusters}...")

# ==============================================================================
# 1. K-MEANS (Base de comparação)
# ==============================================================================
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
# Usa apenas cols_para_treino
labels_kmeans = kmeans.fit_predict(df_scaled[cols_para_treino])

results.append({
    'Modelo': 'K-Means',
    'Silhouette (Maior é melhor)': silhouette_score(df_scaled[cols_para_treino], labels_kmeans),
    'Calinski-Harabasz (Maior é melhor)': calinski_harabasz_score(df_scaled[cols_para_treino], labels_kmeans),
    'Davies-Bouldin (Menor é melhor)': davies_bouldin_score(df_scaled[cols_para_treino], labels_kmeans)
})

# ==============================================================================
# 2. HIERARCHICAL CLUSTERING (Ward Linkage)
# ==============================================================================
hierarchical = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
labels_hc = hierarchical.fit_predict(df_scaled[cols_para_treino])

results.append({
    'Modelo': 'Hierarchical (Ward)',
    'Silhouette (Maior é melhor)': silhouette_score(df_scaled[cols_para_treino], labels_hc),
    'Calinski-Harabasz (Maior é melhor)': calinski_harabasz_score(df_scaled[cols_para_treino], labels_hc),
    'Davies-Bouldin (Menor é melhor)': davies_bouldin_score(df_scaled[cols_para_treino], labels_hc)
})

# ==============================================================================
# 3. GAUSSIAN MIXTURE MODELS (GMM)
# ==============================================================================
gmm = GaussianMixture(n_components=n_clusters, random_state=42)
labels_gmm = gmm.fit_predict(df_scaled[cols_para_treino])

results.append({
    'Modelo': 'GMM (Gaussian Mixture)',
    'Silhouette (Maior é melhor)': silhouette_score(df_scaled[cols_para_treino], labels_gmm),
    'Calinski-Harabasz (Maior é melhor)': calinski_harabasz_score(df_scaled[cols_para_treino], labels_gmm),
    'Davies-Bouldin (Menor é melhor)': davies_bouldin_score(df_scaled[cols_para_treino], labels_gmm)
})

# Criar DataFrame de Comparação
df_results = pd.DataFrame(results).set_index('Modelo')

print("\n--- Tabela de Desempenho Comparativa (Estável) ---")
display(df_results.style.highlight_max(axis=0, color='lightgreen', subset=['Silhouette (Maior é melhor)', 'Calinski-Harabasz (Maior é melhor)'])
                 .highlight_min(axis=0, color='lightgreen', subset=['Davies-Bouldin (Menor é melhor)']))
```

- **Interpretação dos Resultados e Definição do Modelo**

A análise da tabela de desempenho demonstrou a superioridade consistente do **K-Means** frente às abordagens Hierárquica e GMM em todos os critérios avaliados:

Não basta apenas agrupar os dados; é necessário garantir que o agrupamento seja estável, transparente e traduzível para a linguagem de negócio.

Esta etapa foi estruturada nos seguintes processos:

* **Avaliação do modelo aperfeiçoado:** Validação final da coerência dos grupos sob a ótica da geografia econômica e da distribuição espacial.

### **Ajustes dos hiperparâmetros**
A execução inicial do algoritmo K-Means utilizou as configurações padrão (*default*) da biblioteca `scikit-learn`. Embora essas configurações sejam robustas para casos gerais, a aplicação em dados socioeconômicos complexos exige um refinamento técnico para garantir duas condições essenciais:

* **Estabilidade:** Assegurar a reprodutibilidade dos grupos, evitando que a aleatoriedade da inicialização leve a resultados distintos em diferentes execuções (mínimos locais).
* **Convergência:** Garantir que o algoritmo encontre os melhores centróides possíveis para definir cada cluster.

Para atingir esse objetivo, não é utilizada a abordagem empírica de "tentativa e erro", mas sim a técnica de **Grid Search** (Busca em Grade). Este método explora exaustivamente combinações de parâmetros para identificar a configuração que maximiza o coeficiente de Silhueta.

Os seguintes hiperparâmetros foram submetidos à otimização (mantendo-se fixo ):

* **`init`:** Comparação entre a inicialização totalmente aleatória (*random*) e o método *k-means++*, que seleciona centróides iniciais distantes entre si para acelerar a convergência.
* **`n_init`:** Variação do número de vezes que o algoritmo é reiniciado com sementes diferentes (testando valores superiores ao padrão de 10), visando aumentar a probabilidade de encontrar o ótimo global.
* **`algorithm`:** Teste entre as variantes *lloyd* (padrão) e *elkan*, sendo esta última frequentemente mais eficiente computacionalmente em *datasets* densos.

Após a execução dos testes, a configuração que apresentou o melhor equilíbrio estatístico foi:

* _**Melhores Parâmetros**_: {'algorithm': 'lloyd', 'init': 'k-means++', 'max_iter': 300, 'n_init': 10}

* _**Melhor Silhueta:**_ 0.1644

O resultado deste processo permitiu definir a "Melhor Configuração" do modelo, validando se os parâmetros padrão eram suficientes ou se houve ganho incremental de desempenho com o ajuste fino.


- **Grid Search (Busca em Grade)**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score, make_scorer
import numpy as np

# 1. Definimos o modelo base (K=7 fixo)
kmeans_base = KMeans(n_clusters=7, random_state=42)

# 2. Definimos a "Grade" de parâmetros a testar
param_grid = {
    'init': ['k-means++', 'random'],
    'n_init': [10, 20, 50],       # Testar rodar mais vezes para garantir estabilidade
    'max_iter': [300, 500],       # Dar mais tempo para o algoritmo convergir
    'algorithm': ['lloyd', 'elkan'] # Testar variantes do algoritmo
}

# 3. Criamos a métrica de avaliação (Silhueta)
# O GridSearchCV por padrão busca 'score' supervisionado, então precisamos criar um scorer customizado
def silhouette_scorer(estimator, X):
    # Ensure X is a numpy array if it's a pandas DataFrame during scoring
    if hasattr(X, 'values'):
        X = X.values
    labels = estimator.fit_predict(X)
    score = silhouette_score(X, labels)
    return score

# 4. Configuração da Busca (Grid Search)
print("Iniciando Grid Search para Hiperparâmetros (K=7)...")

# Get the number of samples from the data
n_samples = df_scaled.shape[0]
# Create explicit integer index arrays for the single fold
train_idx = np.arange(n_samples)
test_idx = np.arange(n_samples)
custom_cv = [(train_idx, test_idx)]

grid_search = GridSearchCV(
    estimator=kmeans_base,
    param_grid=param_grid,
    scoring=silhouette_scorer,
    cv=custom_cv, # Use the explicitly created integer index arrays instead of slice(None)
    verbose=1,
    n_jobs=-1 # Usar todos os processadores
)

# 5. Execução
grid_search.fit(df_scaled.values) # Convert df_scaled to numpy array

# 6. Resultados
print("\n--- Melhor Configuração Encontrada ---")
print(f"Melhor Silhueta: {grid_search.best_score_:.4f}")
print(f"Melhores Parâmetros: {grid_search.best_params_}")

# Comparação com o modelo anterior (que usou n_init=10, init='k-means++', algorithm='lloyd' por padrão)
```

- **Análise dos Resultados da Otimização**

A execução do *Grid Search* corroborou a eficácia da configuração padrão do algoritmo. Observou-se que a melhor combinação de parâmetros identificada (`n_init=10`, `init='k-means++'`, `algorithm='lloyd'`) replicou exatamente o coeficiente de Silhueta do modelo original.

A ausência de variação métrica não representa um insucesso do ajuste, mas sim uma validação de **robustez**. O resultado demonstra que a estrutura dos clusters formados é estável e bem definida, não dependendo de parametrizações complexas ou de condições específicas de inicialização para ser alcançada. Conclui-se, portanto, que o modelo possui alta confiabilidade técnica para aplicação nos dados propostos.

### **Modelo Final e Geração dos Clusters**

Uma vez determinados os hiperparâmetros ótimos e confirmada a robustez do algoritmo, é realizada a execução definitiva do modelo. O código a seguir realiza a instancialização do algoritmo K-Means configurado com  e a fixação da semente aleatória (`random_state=42`), assegurando a integridade e a total reprodutibilidade dos grupos formados.

O processo de execução compreende três ações centrais:

1. **Treinamento:** O modelo processa as variáveis normalizadas, identificando os centróides matemáticos que melhor segmentam o território nacional.
2. **Atribuição (*Labeling*):** Cada um dos 5.570 municípios recebe um rótulo correspondente ao seu perfil econômico, que é anexado ao *dataframe* analítico.
3. **Visualização dos Perfis:** Geração de um mapa de calor (*heatmap*) das médias padronizadas. Esta visualização é a ferramenta importante pois permite observar, através da intensidade das cores, quais características (ex: PIB, Emprego Industrial, Dinâmica de Serviços) são dominantes ou deficitárias em cada cluster

A execução do modelo final identificou **7 clusters** e validou a robustez da segmentação, demonstrando alta sensibilidade ao isolar outliers de magnitude econômica e distinguir nichos vocacionais específicos da base municipal predominante.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ==============================================================================
# 1. INSTANCIAÇÃO E TREINAMENTO DO MODELO FINAL (CORRIGIDO)
# ==============================================================================

# Garante a seleção APENAS das colunas numéricas (Features), ignorando a coluna 'cluster' se ela existir
# Isso previne que o resultado anterior influencie a nova execução
cols_para_treino = [col for col in df_scaled.columns if col != 'cluster']

kmeans_final = KMeans(
    n_clusters=7,
    init='k-means++',
    n_init=10,
    max_iter=300,
    algorithm='lloyd',
    random_state=42 # Fixa a semente matemática
)

# Treina apenas com as colunas limpas
clusters_final = kmeans_final.fit_predict(df_scaled[cols_para_treino])

# ==============================================================================
# 2. ATRIBUIÇÃO DOS CLUSTERS (LABELING)
# ==============================================================================
# Cria uma cópia fresca para o dataframe final
df_ml_final = df_ml_corr.copy()

# Atribui os labels
df_ml_final.loc[:, 'cluster'] = clusters_final
df_scaled.loc[:, 'cluster'] = clusters_final

# ==============================================================================
# 3. APRESENTAÇÃO DOS RESULTADOS
# ==============================================================================

# A. Número de Municípios por Cluster
print("--- Distribuição Final dos Municípios por Cluster ---")
distribuicao = df_ml_final['cluster'].value_counts().sort_index()
display(distribuicao)
print(f"Total de Municípios classificados: {distribuicao.sum()}")

# B. Perfil dos Clusters (Heatmap)
# Agrupa pela coluna cluster
cluster_profiles = df_scaled.groupby('cluster')[cols_para_treino].mean()

# Plotagem
plt.figure(figsize=(18, 10))
sns.heatmap(cluster_profiles.T,
            cmap='RdBu_r',
            center=0,
            annot=True,
            fmt=".2f",
            linewidths=.5)

plt.title('Identidade dos 7 Clusters (Estável e Reprodutível)', fontsize=16, pad=20)
plt.xlabel('Cluster (ID)', fontsize=12)
plt.ylabel('Features (Variáveis)', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### **Explicabilidade**

A explicabilidade em *Machine Learning* refere-se à capacidade de tornar transparente o funcionamento interno de um modelo, permitindo compreender as razões lógico-matemáticas que determinaram um agrupamento específico. No contexto de políticas públicas, esta etapa é fundamental, pois converte abstrações algorítmicas em critérios objetivos e auditáveis, essenciais para a tomada de decisão e a transparência institucional.

Para interpretar a lógica de segmentação do algoritmo K-Means, três técnicas complementares de investigação são aplicadas:

Esta técnica de "caixa branca" permite extrair as regras de negócio explícitas (ex: "Se PIB > X e Saldo < Y, então Cluster Z") que mimetizam o comportamento do algoritmo principal.
* **Análise de Distribuição:** Visualização da dispersão das variáveis via *boxplots*, utilizada para validar a separação entre os grupos e identificar as faixas de valores (*ranges*) que caracterizam cada perfil econômico.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text

# ==============================================================================
# CONFIGURAÇÃO INICIAL
# ==============================================================================
# Lista completa de todas as variáveis numéricas originais do projeto
features_originais = [
    'pib', 'vl_agro', 'vl_ind', 'vl_adm',
    'est_mes_total', 'est_mes_agro', 'est_mes_ind',
    'saldo_agro', 'saldo_ind', 'saldo_com', 'saldo_serv', 'saldo_const',
    'vr_rel_total', 'tp_emp_total', 'tp_emp_agro', 'tp_emp_ind'
]

# ==============================================================================
# PASSO 1: TABELA DE CENTRÓIDES (PERFIL MÉDIO/MEDIANO)
# ==============================================================================
print("--- Passo 1: Perfil Típico dos Clusters (Mediana) ---")

# Agrupando pela Mediana
perfil_mediana = df_ml_final.groupby('cluster')[features_originais].median()
display(perfil_mediana.round(2))

# ==============================================================================
# PASSO 2: MODELO SUBSTITUTO (ÁRVORE DE DECISÃO)
# ==============================================================================
print("\n--- Passo 2: Regras de Negócio (Árvore de Decisão) ---")

tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(df_ml_final[features_originais], df_ml_final['cluster'])

# Plotagem Gráfica
plt.figure(figsize=(25, 12))
plot_tree(tree_model,
          feature_names=features_originais,
          class_names=[str(c) for c in sorted(df_ml_final['cluster'].unique())],
          filled=True, rounded=True, fontsize=10)
plt.title("Árvore de Decisão Explicativa", fontsize=16)
plt.show()

# Regras em Texto
print("Regras de Decisão Extraídas:")
print(export_text(tree_model, feature_names=features_originais))

# ==============================================================================
# PASSO 3: VISUALIZAÇÃO DE DISTRIBUIÇÃO (TODAS AS VARIÁVEIS) - CORRIGIDO
# ==============================================================================
print("\n--- Passo 3: Distribuição e Separação dos Grupos (Boxplots Completos) ---")

variaveis_para_plotar = features_originais

# Ajuste dinâmico da altura: 3 polegadas de altura por gráfico
plt.figure(figsize=(15, 3 * len(variaveis_para_plotar)))

for i, col in enumerate(variaveis_para_plotar):
    plt.subplot(len(variaveis_para_plotar), 1, i + 1)

    # CORREÇÃO APLICADA AQUI:
    # Adicionado hue='cluster' e legend=False para atender ao FutureWarning do Seaborn
    sns.boxplot(x='cluster', y=col, hue='cluster', data=df_ml_final, palette="Set3", showfliers=False, legend=False)

    plt.title(f'Distribuição de {col} por Cluster', fontsize=11, fontweight='bold')
    plt.ylabel(col)
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

A estratégia de implantação prioriza a acessibilidade e a visualização para os tomadores de decisão, superando a barreira técnica do código. Como os dados demográficos não exigem processamento em tempo real (baixa latência), a arquitetura de entrega é desenhada em níveis de maturidade progressiva:

* **Nível 1: Painéis de Inteligência de Negócio (BI):** Integração da base final classificada (`Painel_Estrategico_IFs.csv`) com ferramentas corporativas de visualização (ex: Power BI, Tableau ou Looker). * **Nível 2: Aplicação Web Interativa:** Desenvolvimento de uma interface amigável (via *frameworks* como Streamlit) que permita consultas granulares. Nesta camada, o usuário insere o município e o sistema retorna um relatório automático contendo o perfil econômico, a comparação com vizinhos regionais e a recomendação de portfólio de cursos.
* **Nível 3: Integração Sistêmica (ERP):** Incorporação dos atributos `Perfil_Economico` e `Eixo_Prioritario` diretamente nas bases de dados dos sistemas acadêmicos institucionais (como o SUAP). Isso permite a criação de "alertas de governança": ao cadastrar um novo curso, o sistema pode cruzar a intenção de oferta com a vocação local identificada, sugerindo justificativas caso haja divergência.

### **Monitorar (*Monitoring*)**

A validade de um modelo socioeconômico é temporalmente sensível. Fatores como crises econômicas, novas legislações ou choques de oferta podem alterar a realidade dos municípios. O plano de monitoramento visa detectar a degradação do modelo através de três dimensões:

* **Deriva de Dados (*Data Drift*):** Verificação anual das distribuições estatísticas das variáveis. Mudanças estruturais (ex: inflação descontrolada afetando o PIB nominal) exigem intervenções de pré-processamento (deflacionamento) para evitar que o modelo interprete erroneamente o crescimento monetário como desenvolvimento real.
* **Estabilidade dos Clusters:** Análise da taxa de migração dos municípios entre grupos após novas cargas de dados. Uma taxa de mudança marginal (5-10%) é esperada e reflete a evolução natural das cidades. Contudo, taxas elevadas (ex: >40%) indicam perda de validade das fronteiras de decisão, sinalizando a necessidade de re-treinamento imediato.
* **Monitoramento de Eficácia (Feedback de Negócio):** Estabelecimento de indicadores de sucesso (KPIs) baseados na política pública, como a "Taxa de Ocupação de Vagas" e "Empregabilidade de Egressos". O objetivo é validar se os cursos abertos com base na recomendação do modelo apresentam melhor desempenho do que as ofertas definidas sem base analítica.

### **Manter (*Maintenance*)**

A manutenção assegura a longevidade da ferramenta frente à atualização das fontes primárias (IBGE, CAGED, RAIS), que ocorre periodicamente. O protocolo de manutenção define:

* **Recalibragem Anual (*Retraining*):** Execução do *pipeline* de treinamento com os dados mais recentes, mantendo a padronização (`StandardScaler`). É crucial realizar a reanálise semântica dos grupos, pois a numeração matemática dos clusters (0, 1, 2...) pode alterar-se a cada execução, exigindo nova validação humana para garantir que o "Cluster Agro" continue mapeado corretamente.
* **Versionamento de Modelos:** Manutenção de um histórico de versões (`Modelo_v2024`, `Modelo_v2025`). Isso permite importantes análises longitudinais: se um município migrou do perfil "Economia de Subsistência" para "Economia Mista" ao longo dos anos, essa transição torna-se um indicador de impacto do desenvolvimento regional.
* **Documentação Viva:** Atualização contínua do dicionário de dados e das regras de negócio. A inclusão de novas variáveis explicativas (ex: infraestrutura de conectividade 5G) deve seguir rigoroso processo de teste e documentação antes de integrar o modelo produtivo.

## **Preparação dos dados para as análises**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================================================================
# 0. PREPARAÇÃO DOS DADOS DE OFERTA (df_et)
# ==============================================================================
print("--- Passo 0: Processando Dados de Oferta Real ---")

# Garantindo que cod_ibge seja do mesmo tipo em ambos (string ou int) para o merge funcionar
df_et['cod_ibge'] = df_et['cod_ibge'].astype(int)
painel_decisao['cod_ibge'] = painel_decisao['cod_ibge'].astype(int)

# Descobrindo o Eixo Dominante (com mais vagas) em cada município
# Agrupa por cidade e eixo, soma as vagas
oferta_por_cidade = df_et.groupby(['cod_ibge', 'eixo'])['vagas'].sum().reset_index()

# Ordena para pegar o eixo com mais vagas por cidade
oferta_dominante = oferta_por_cidade.sort_values(['cod_ibge', 'vagas'], ascending=[True, False])
oferta_dominante = oferta_dominante.drop_duplicates(subset=['cod_ibge'], keep='first')

# Renomeia para facilitar o merge
oferta_dominante = oferta_dominante.rename(columns={'eixo': 'Eixo_Real_Ofertado', 'vagas': 'Vagas_Reais'})

# CRUZA COM A INTELIGÊNCIA DO MODELO
# Merge Left: Mantemos todos os 5570 municípios do painel. Quem não tiver oferta ficará com NaN/Nulo.
df_confronto = pd.merge(painel_decisao, oferta_dominante[['cod_ibge', 'Eixo_Real_Ofertado', 'Vagas_Reais']],
                        on='cod_ibge', how='left')

# Cria flag de status: "Com Oferta" ou "Sem Oferta"
df_confronto['Status_Oferta'] = df_confronto['Eixo_Real_Ofertado'].apply(lambda x: 'Sem Oferta IF' if pd.isna(x) else 'Com Oferta IF')

print(f"Base Cruzada Pronta! Total de Municípios: {len(df_confronto)}")
print(f"Municípios com oferta identificada: {df_confronto['Eixo_Real_Ofertado'].notnull().sum()}")
```

## **Aplicação 1: Diagnóstico de Aderência da Oferta**
Esta aplicação tem por objetivo mensurar o grau de alinhamento estratégico entre a oferta educacional vigente nos Institutos Federais e as vocações econômicas reais identificadas pelo modelo. A metodologia consiste em um teste lógico de correspondência (*match flexível*): considera-se que há aderência quando o Eixo Tecnológico com maior volume de oferta de vagas no município coincide com o **Eixo Prioritário 1** ou com o **Eixo Prioritário 2** sugeridos pelo algoritmo de *clustering*. Esta métrica binária permite calcular a Taxa de Aderência, servindo como indicador de eficiência na alocação de vagas.

---

**Resumo do resultado da análise:**

Em resumo a análise diagnóstica revelou um **desalinhamento estratégico crítico**, evidenciado por uma **Taxa Global de Aderência de apenas 13,7%**, o que indica que em **86,3% dos municípios a oferta educacional predominante diverge da vocação econômica local**. Regionalmente, o Centro-Oeste apresentou o melhor desempenho (16,7%) devido à clareza do perfil agropecuário, enquanto o Sul registrou o índice mais baixo (9,8%), refletindo lacunas no atendimento a cadeias industriais complexas. **Os dados sugerem que a definição de cursos foi historicamente guiada pela inércia institucional e pela preferência por formações de baixo custo infraestrutural (como Gestão)**, em detrimento de áreas técnicas especializadas que demandam maior investimento.

```python
# ==============================================================================
# APLICAÇÃO 1: CONFRONTO (ADERÊNCIA DA OFERTA)
# ==============================================================================
print("\n" + "="*80)
print("APLICAÇÃO 1: DIAGNÓSTICO DE ADERÊNCIA (O que existe vs. O que deveria existir)")
print("="*80)

# Filtrar apenas quem tem oferta
df_analise_oferta = df_confronto[df_confronto['Eixo_Real_Ofertado'].notnull()].copy()

# Lógica de Match Flexível (Acerto = Oferta bate com Eixo 1 OU Eixo 2)
def verificar_aderencia(row):
    eixo_real = row['Eixo_Real_Ofertado']
    # Normaliza textos (maiusculas/minusculas) para evitar erro bobo de string
    if pd.isna(eixo_real): return False
    return (eixo_real == row['Eixo_Prioritario_1']) or (eixo_real == row['Eixo_Prioritario_2'])

df_analise_oferta['Tem_Aderencia'] = df_analise_oferta.apply(verificar_aderencia, axis=1)

# Cálculo da Taxa de Aderência por Região
taxa_aderencia = df_analise_oferta.groupby('regiao')['Tem_Aderencia'].mean() * 100
taxa_aderencia = taxa_aderencia.sort_values(ascending=False)

# Visualização
plt.figure(figsize=(10, 5))
ax = sns.barplot(x=taxa_aderencia.index, y=taxa_aderencia.values, hue=taxa_aderencia.index, palette='RdYlGn', legend=False)
plt.title('Taxa de Aderência da Oferta Atual à Vocação Econômica (Região)', fontsize=14)
plt.ylabel('% de Oferta Alinhada (Eixo 1 ou 2)')
plt.ylim(0, 100) # Escala de 0 a 100%

# Adiciona os rótulos nas barras
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f%%', padding=3)

plt.show()
print(f"Taxa Global de Aderência Brasil: {df_analise_oferta['Tem_Aderencia'].mean()*100:.1f}%")
```

- **Análise do Resultado**

Os resultados obtidos através do confronto entre a base de dados de vagas e os *clusters* econômicos revelam um cenário de desalinhamento estrutural que valida a importância da abordagem orientada a dados.

**1. Desalinhamento Estratégico Global**
A Taxa Global de Aderência apurada foi de apenas **13,7%**. Estatisticamente, isso indica que em **86,3%** dos municípios analisados, o principal eixo ofertado não dialoga com as vocações econômicas primárias ou secundárias do território. Este indicador sugere que, historicamente, a abertura de cursos pode ter sido orientada por critérios de conveniência infraestrutural ou demandas pregressas, descolando-se da realidade produtiva atual medida pelo PIB e estoque de empregos.

**2. Assimetrias Regionais**
A análise estratificada por região expõe nuances na eficiência da oferta:

* **Centro-Oeste (Maior Aderência - 16,7%):** O desempenho superior pode ser atribuído à clareza da vocação regional. A predominância inequívoca do agronegócio facilita o planejamento pedagógico, induzindo naturalmente a oferta de cursos no eixo de Recursos Naturais, o que gera o *match* com o modelo.
* **Sul (Menor Aderência - 9,8%):** A região apresenta o maior hiato de oferta. A hipótese analítica é que a complexidade econômica do Sul (com clusters industriais e diversificados) não está sendo atendida por cursos específicos. É provável que haja uma predominância de ofertas genéricas (como Gestão e Negócios) em territórios que demandam formação industrial especializada (Metalmecânica, Automação), gerando a baixa pontuação da taxa de aderência.

**3. Fatores Explicativos do Desalinhamento**
Três hipóteses principais fundamentam os baixos índices encontrados:

* **Custo de Implementação (*Trade-off* Infraestrutural):** Eixos como "Gestão e Negócios" e "Informação e Comunicação" possuem baixo custo de implantação. Em contraste, eixos de "Controle e Processos Industriais" ou "Saúde" exigem laboratórios complexos. A gestão pode ter optado pela expansão via cursos de baixo custo, ignorando vocações locais mais onerosas.
* **Inércia Institucional:** Existe uma latência entre a mudança econômica e a atualização do catálogo de cursos. Campi abertos há décadas podem manter ofertas tradicionais (ex: Agropecuária) enquanto a cidade evoluiu para um perfil de Serviços ou Comércio, criando um anacronismo na oferta.
* **Metodologia de Volume:** O indicador avalia a *oferta principal*. É possível que o *campus* oferte o curso vinculado ao eixo tecnológico indicado pelo modelo (prioritário 1 ou 2), mas com poucas vagas, enquanto prioriza massivamente cursos genéricos, o que penaliza sua taxa de aderência nesta métrica específica.

**Síntese para Governança**
O diagnóstico de 13,7% de aderência evidencia um ponto crítico de gestão: há um volume massivo de recursos públicos alocados em formações que não atacam a principal alavanca econômica dos municípios. A correção deste desvio representa a maior oportunidade de ganho de eficiência institucional mapeada por este estudo.

## **Aplicação 2: Análise de Divergência Total**
Diferente da aplicação anterior, que analisava a predominância da oferta, este indicador adota um critério de avaliação mais abrangente e "generoso". A metodologia verifica se **algum** dos eixos tecnológicos presentes no portfólio do *campus* (independentemente do número de vagas) corresponde à vocação econômica primária ou secundária identificada pelo modelo.
O objetivo é identificar os casos de **Divergência Total**: municípios onde a instituição oferta cursos técnicos, mas **nenhum deles** dialoga com os motores econômicos locais. Este indicador revela a existência de "ilhas acadêmicas", onde a formação técnica ocorre de costas para a realidade produtiva do território.

---

**Resumo do resultado da análise:**

Em resumo, os resultados da análise revelam um desalinhamento estrutural crítico, evidenciado por uma Taxa de Divergência Nacional de 44,4%, o que indica que quase metade dos municípios atendidos não oferta sequer um curso alinhado às vocações econômicas locais (PIB e Emprego). Regionalmente, o Centro-Oeste apresenta a maior polarização (50,0% de divergência), enquanto estados como Mato Grosso (72,2%) e Paraná (61,5%) figuram como casos sistêmicos, onde a oferta acadêmica ignora demandas produtivas evidentes, como a base industrial paranaense. Este cenário de "ilhas acadêmicas" exige uma revisão imediata dos Planos de Desenvolvimento Institucional (PDIs), priorizando a correção de matrizes de oferta que hoje induzem à evasão profissional e desperdício de eficiência pública.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================================================================
# 1. PREPARAÇÃO DOS DADOS (LÓGICA "QUALQUER EIXO")
# ==============================================================================
print("--- Processando Oferta Total vs. Vocação ---")

# Normalização de Tipos
df_et['cod_ibge'] = df_et['cod_ibge'].astype(int)
painel_decisao['cod_ibge'] = painel_decisao['cod_ibge'].astype(int)

# EM VEZ DE PEGAR O MÁXIMO, AGORA PEGAMOS TODOS OS EIXOS OFERTADOS
# Cria uma lista de eixos únicos por município
oferta_por_cidade = df_et.groupby('cod_ibge')['eixo'].apply(set).reset_index()
oferta_por_cidade = oferta_por_cidade.rename(columns={'eixo': 'Lista_Eixos_Ofertados'})

# Merge com o Painel de Decisão
df_confronto = pd.merge(painel_decisao, oferta_por_cidade, on='cod_ibge', how='left')

# Filtra apenas quem tem oferta (para calcular a taxa corretamente)
df_analise_oferta = df_confronto[df_confronto['Lista_Eixos_Ofertados'].notnull()].copy()

print(f"Total de Municípios com Oferta IF avaliados: {len(df_analise_oferta)}")

# ==============================================================================
# 2. CÁLCULO DA DIVERGÊNCIA (NOVA LÓGICA)
# ==============================================================================

def verificar_divergencia_total(row):
    eixos_cidade = row['Lista_Eixos_Ofertados'] # Isso agora é um conjunto (set) de eixos

    prio_1 = row['Eixo_Prioritario_1']
    prio_2 = row['Eixo_Prioritario_2']

    # É aderente se Prio 1 está na lista OU Prio 2 está na lista
    eh_aderente = (prio_1 in eixos_cidade) or (prio_2 in eixos_cidade)

    # Divergência é o oposto de aderência
    return not eh_aderente

# Aplica a verificação
df_analise_oferta['Eh_Divergente'] = df_analise_oferta.apply(verificar_divergencia_total, axis=1)

# Cálculo das Taxas (Média de True/False * 100)
# (Nº Divergentes / Nº Total com Oferta)
taxa_div_regiao = df_analise_oferta.groupby('regiao')['Eh_Divergente'].mean() * 100
taxa_div_uf = df_analise_oferta.groupby('uf_sigla')['Eh_Divergente'].mean() * 100

# Ordenação para o gráfico
taxa_div_regiao = taxa_div_regiao.sort_values(ascending=False)
taxa_div_uf = taxa_div_uf.sort_values(ascending=False)

# ==============================================================================
# 3. VISUALIZAÇÃO CORRIGIDA
# ==============================================================================

# Gráfico 1: Divergência por Região
plt.figure(figsize=(15, 5))
ax = sns.barplot(x=taxa_div_regiao.index, y=taxa_div_regiao.values, hue=taxa_div_regiao.index, legend=False, palette='Reds_r')
plt.title('Taxa de Divergência por região (Nenhum eixo ofertado atende a vocação)', fontsize=14)
plt.ylabel('% de Municípios Divergentes')
plt.ylim(0, 100)
for i in ax.containers: ax.bar_label(i, fmt='%.1f%%', padding=3)
plt.show()

# Gráfico 2: Ranking de Divergência por UF
plt.figure(figsize=(20, 6))
top_divergentes = taxa_div_uf
ax2 = sns.barplot(x=top_divergentes.index, y=top_divergentes.values, hue=top_divergentes.index, legend = False, palette='Reds_r')
plt.title('Taxa de Divergência por estado (Oferta vs. Modelo)', fontsize=14)
plt.ylabel('% de Municípios Divergentes')
plt.ylim(0, 100)
for i in ax2.containers: ax2.bar_label(i, fmt='%.1f%%', padding=3)
plt.show()

media_nacional = df_analise_oferta['Eh_Divergente'].mean() * 100
print(f"Taxa de Divergência Nacional (Nova Lógica): {media_nacional:.1f}%")
print(f"Isso significa que em {media_nacional:.1f}% das cidades com IF, NENHUM eixo ofertado bate com eixo prioritário 1 ou 2.")

# ==============================================================================
# 4. LISTA DE DIVERGÊNCIA (QUEM SÃO ELES?)
# ==============================================================================
print("\n" + "="*80)
print("RELATÓRIO DE MUNICÍPIOS TOTALMENTE DESALINHADOS")
print("="*80)

# Filtrando
divergentes_final = df_analise_oferta[df_analise_oferta['Eh_Divergente']].copy()

# Ajustando a coluna de lista para ficar legível no dataframe (tirar de set para string)
divergentes_final['Lista_Eixos_Ofertados'] = divergentes_final['Lista_Eixos_Ofertados'].apply(lambda x: ", ".join(x))

cols_view = ['cod_ibge', 'cidade', 'uf_sigla', 'Perfil_Economico',
             'Lista_Eixos_Ofertados', 'Eixo_Prioritario_1', 'Eixo_Prioritario_2']

print(f"Total de Casos Críticos: {len(divergentes_final)}")
display(divergentes_final[cols_view].head(15))
```
